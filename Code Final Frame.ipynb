```python
#new Kalman filter method 
#the Kalman function to filter the ENVIRONMENTAL EXPONENTIAL from the previous ENVIRONMENTAL EXPONENTIAL
import numpy as np
import matplotlib.pyplot as plt

# Define the Kalman filter parameters and initial values
A = 1
H = 1

#system and measurement error variables
Q = 0
R = 0.16 #take the square of standard deviation

#the initial values and the error of that
x = 2.1 #take the average 
P = 10 #4
firstRun = True

def SimpleKalman(z):
    global A, H, Q, R, x, P, firstRun

    if firstRun:
        firstRun = False

    # Kalman algorithm
    xp = A * x  # I. Prediction of estimate
    Pp = A * P * A + Q  # Prediction of error cov

    K = Pp * H / (H * Pp * H + R)  # II. Computation of Kalman gain

    x = xp + K * (z - H * xp)  # III. Comp. of state estimate
    P = Pp - K * H * Pp  # IV. Comp. of error cov.

    return x

filteredna_values = []
# Simulation parameters
dt = 0.2  # time step (s)
t = np.arange(0, len(na_values) * dt, dt)  # Create an array for time
Nsamples = len(t)

Xsaved = np.zeros(Nsamples)
Zsaved = np.zeros(Nsamples)

# Simulate Kalman filter with the provided dataset
for k in range(Nsamples):
    z = na_values[k]
    filteredRSSI = SimpleKalman(z)

    Xsaved[k] = filteredRSSI
    Zsaved[k] = z
    filteredna_values.append(filteredRSSI)

# Create a line plot
fig, ax = plt.subplots(figsize=(10, 6), facecolor='white')
t = range(Nsamples)  # Generate a simple time axis
ax.plot(t, Xsaved, 'o-',markersize=7, linewidth=2, label='Kalman Filter')
ax.plot(t, Zsaved, 'r:*', markersize=7, linewidth=2, label='Measurements')
ax.set_xlabel('Sample')
ax.set_ylabel('Path loss exponent')
ax.set_title('Kalman Filter Path loss exponent')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

# Save the plot with a white background
fig.savefig("rssi_kalman_filter_path_loss_exponent.png", dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
